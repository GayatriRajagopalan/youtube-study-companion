# YouTube Study Companion

### RAG-powered learning assistant for long educational YouTube videos

This notebook is a useful RAG application that helps learners study long YouTube videos.

## Core features in this version

1. Extract YouTube transcript
2. Create transcript chunks
3. Generate OpenAI embeddings
4. Store chunks in ChromaDB
5. Ask questions with timestamped answers
6. Extract topic timeline
7. Generate study pack
8. Generate quiz
9. Use a clean Gradio UI

# Step 1: Setup

```python
APP_CONFIG = {
    "embedding_model": "text-embedding-3-small",
    "chat_model": "gpt-4o-mini",
    "batch_size": 10,
    "chunk_duration": 300,
    "db_name": "youtube_learning_db",
    "collection_name": "youtube_study_chunks",
    "top_k": 3,
    "max_transcript_chars": 12000
}
```

```python
import os
import sys

from dotenv import load_dotenv
from openai import OpenAI
from chromadb import PersistentClient
from youtube_transcript_api import YouTubeTranscriptApi
from pydantic import BaseModel
import gradio as gr
```

```python
print("Python executable:", sys.executable)

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")

if api_key:
    print("API key loaded successfully")
else:
    print("API key missing. Please check your .env file.")

openai = OpenAI()
```

```python
client = PersistentClient(
    path=APP_CONFIG["db_name"]
)

collection = client.get_or_create_collection(
    APP_CONFIG["collection_name"]
)

embedding_model = APP_CONFIG["embedding_model"]
chat_model = APP_CONFIG["chat_model"]
```

# Step 2: Pydantic Models

```python
class VideoInfo(BaseModel):
    video_id: str
    title: str | None = None
    channel: str | None = None
    url: str | None = None


class TranscriptSegment(BaseModel):
    video_id: str
    text: str
    start_time: float
    duration: float

    @property
    def end_time(self) -> float:
        return self.start_time + self.duration


class TranscriptChunk(BaseModel):
    video_id: str
    chunk_id: str
    text: str
    start_time: float
    end_time: float


class RetrievedChunk(BaseModel):
    chunk_id: str
    text: str
    start_time: float
    end_time: float
    score: float | None = None


class Topic(BaseModel):
    title: str
    description: str
    start_time: str


class TopicList(BaseModel):
    topics: list[Topic]


class StudyPack(BaseModel):
    topic: str
    explanation: str
    key_points: list[str]
    example: str
    interview_style_summary: str


class QuizQuestion(BaseModel):
    question: str
    options: list[str]
    answer: str
    explanation: str


class Quiz(BaseModel):
    questions: list[QuizQuestion]
```

# Step 3: Transcript Fetching

```python
def get_video_id(video_url: str) -> str:

    # This simple version supports normal YouTube links like:
    # https://www.youtube.com/watch?v=TljgQ2ZRT1U

    if "v=" in video_url:
        video_id = video_url.split("v=")[1]
        video_id = video_id.split("&")[0]
        return video_id

    # If the user directly pastes only the video id
    return video_url.strip()
```

```python
def fetch_transcript(video_url: str):

    video_id = get_video_id(video_url)

    ytt_api = YouTubeTranscriptApi()

    transcript = ytt_api.fetch(video_id)

    segments = []

    for item in transcript:

        segment = TranscriptSegment(
            video_id=video_id,
            text=item.text,
            start_time=item.start,
            duration=item.duration
        )

        segments.append(segment)

    return segments
```

# Step 4: Chunk Creation

```python
def create_chunks(segments):

    chunks = []

    chunk_text = ""

    chunk_start = None
    chunk_end = None

    CHUNK_DURATION = APP_CONFIG["chunk_duration"]

    chunk_number = 0

    for segment in segments:

        start_time = segment.start_time
        end_time = segment.end_time
        text = segment.text

        if chunk_start is None:
            chunk_start = start_time

        chunk_text += " " + text

        chunk_end = end_time

        if chunk_end - chunk_start >= CHUNK_DURATION:

            chunk = TranscriptChunk(
                video_id=segment.video_id,
                chunk_id=f"{segment.video_id}_{chunk_number}",
                text=chunk_text.strip(),
                start_time=chunk_start,
                end_time=chunk_end
            )

            chunks.append(chunk)

            chunk_text = ""
            chunk_start = None
            chunk_end = None
            chunk_number += 1

    # Add remaining transcript as the final chunk

    if chunk_text:

        chunk = TranscriptChunk(
            video_id=segments[0].video_id,
            chunk_id=f"{segments[0].video_id}_{chunk_number}",
            text=chunk_text.strip(),
            start_time=chunk_start,
            end_time=chunk_end
        )

        chunks.append(chunk)

    return chunks
```

# Step 5: Embeddings and ChromaDB Storage

```python
def create_embeddings(texts):

    response = openai.embeddings.create(
        model=embedding_model,
        input=texts
    )

    embeddings = [
        item.embedding
        for item in response.data
    ]

    return embeddings
```

```python
def store_chunks(chunks):

    global collection

    # Clear the old collection so the app focuses on one video at a time.
    # This keeps the beginner version simple and avoids mixing videos.

    try:
        client.delete_collection(
            APP_CONFIG["collection_name"]
        )
    except Exception:
        pass

    collection = client.get_or_create_collection(
        APP_CONFIG["collection_name"]
    )

    texts = [chunk.text for chunk in chunks]

    metadatas = [
        {
            "video_id": chunk.video_id,
            "start_time": chunk.start_time,
            "end_time": chunk.end_time
        }
        for chunk in chunks
    ]

    ids = [chunk.chunk_id for chunk in chunks]

    for i in range(0, len(texts), APP_CONFIG["batch_size"]):

        batch_texts = texts[i:i + APP_CONFIG["batch_size"]]
        batch_metadatas = metadatas[i:i + APP_CONFIG["batch_size"]]
        batch_ids = ids[i:i + APP_CONFIG["batch_size"]]

        batch_embeddings = create_embeddings(batch_texts)

        collection.add(
            ids=batch_ids,
            documents=batch_texts,
            metadatas=batch_metadatas,
            embeddings=batch_embeddings
        )
```

```python
def process_video(video_url: str):

    segments = fetch_transcript(video_url)

    chunks = create_chunks(segments)

    store_chunks(chunks)

    return segments, chunks
```

# Step 6: Retrieval

```python
def retrieve_chunks(question: str, top_k: int = 3):

    response = openai.embeddings.create(
        model=embedding_model,
        input=question
    )

    question_embedding = response.data[0].embedding

    results = collection.query(
        query_embeddings=[question_embedding],
        n_results=top_k
    )

    retrieved_chunks = []

    if not results["ids"] or not results["ids"][0]:
        return retrieved_chunks

    for i in range(len(results["ids"][0])):

        chunk = RetrievedChunk(
            chunk_id=results["ids"][0][i],
            text=results["documents"][0][i],
            start_time=results["metadatas"][0][i]["start_time"],
            end_time=results["metadatas"][0][i]["end_time"],
            score=results["distances"][0][i]
        )

        retrieved_chunks.append(chunk)

    return retrieved_chunks
```

# Step 7: RAG Question Answering

```python
def format_timestamp(seconds: float) -> str:

    hours = int(seconds // 3600)

    minutes = int((seconds % 3600) // 60)

    remaining_seconds = int(seconds % 60)

    return f"{hours:02d}:{minutes:02d}:{remaining_seconds:02d}"
```

```python
def build_context(retrieved_chunks, include_timestamps: bool = False):

    retrieved_chunks.sort(key=lambda x: x.start_time)

    context = ""

    for chunk in retrieved_chunks:

        if include_timestamps:
            context += f"""
Timestamp: {format_timestamp(chunk.start_time)} - {format_timestamp(chunk.end_time)}
Transcript:
{chunk.text}
"""
        else:
            context += f"""
Transcript:
{chunk.text}
"""

    return context
```

```python
def answer_question(question: str, top_k: int = 3):

    retrieved = retrieve_chunks(question, top_k=top_k)

    if not retrieved:
        return "Please process a video first, or try asking a different question."

    context = build_context(retrieved, include_timestamps=False)

    prompt = f"""
You are a helpful learning assistant.

Use only the transcript context below to answer the user's question.
Explain the answer in simple language.
Do not mention timestamps in the answer.
If the answer is not available in the context, say that clearly.

Question:
{question}

Transcript context:
{context}
"""

    response = openai.chat.completions.create(
        model=chat_model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.2
    )

    answer = response.choices[0].message.content

    return answer
```

# Step 8: Topic Timeline

```python
def extract_topics_structured():

    all_chunks = collection.get()

    transcript = ""

    for doc, metadata in zip(all_chunks["documents"], all_chunks["metadatas"]):

        transcript += f"""
Timestamp: {format_timestamp(metadata["start_time"])}

Transcript:
{doc}

"""

   

    prompt = f"""
Analyze this educational video transcript.

Extract the major learning topics in the order they appear.

For each topic provide:
- topic title
- short description
- approximate timestamp

Use the timestamp shown before each transcript section.
Do not create too many tiny topics.
Focus on the main sections a learner would care about.

Transcript:

{transcript}
"""

    response = openai.beta.chat.completions.parse(
        model=chat_model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        response_format=TopicList,
        temperature=0.2
    )

    return response.choices[0].message.parsed
```

```python
def show_topics(topic_list: TopicList):

    for topic in topic_list.topics:
        print(f"{topic.start_time} - {topic.title}")
        print(topic.description)
        print()
```

# Step 9: Study Pack

```python
def generate_study_pack(topic_name: str):

    question = f"Explain the topic: {topic_name}"

    retrieved = retrieve_chunks(
        question,
        top_k=3
    )

    if not retrieved:
        return None

    context = build_context(retrieved, include_timestamps=False)

    prompt = f"""
Create a beginner-friendly study pack for the topic: {topic_name}

Use only the transcript context below.

Return:
- topic
- explanation
- key_points
- example
- interview_style_summary

Do not include timestamps.
The interview_style_summary should sound simple, clear, and professional.

Transcript context:
{context}
"""

    response = openai.beta.chat.completions.parse(
        model=chat_model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        response_format=StudyPack,
        temperature=0.2
    )

    return response.choices[0].message.parsed
```

# Step 10: Quiz

```python
def generate_quiz(topic_name: str):

    question = f"Create a quiz about the topic: {topic_name}"

    retrieved = retrieve_chunks(
        question,
        top_k=3
    )

    if not retrieved:
        return None

    context = build_context(retrieved, include_timestamps=False)

    prompt = f"""
Create 5 beginner-friendly multiple choice questions.

Topic:
{topic_name}

Use only the transcript context below.

For each question provide:
- question
- four options
- correct answer
- short explanation

Do not include timestamps.

Transcript context:
{context}
"""

    response = openai.beta.chat.completions.parse(
        model=chat_model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        response_format=Quiz,
        temperature=0.2
    )

    return response.choices[0].message.parsed
```

```python
def show_quiz(quiz: Quiz):

    for i, question in enumerate(quiz.questions, start=1):

        print(f"Question {i}: {question.question}")

        for option in question.options:
            print(f"- {option}")

        print(f"Answer: {question.answer}")
        print(f"Explanation: {question.explanation}")
        print()
```

# Step 11: Gradio Helper Functions

```python
def process_video_for_ui(video_url: str):

    try:
        segments, chunks = process_video(video_url)

        return (
            f"Video processed successfully.\n"
            f"Transcript segments: {len(segments)}\n"
            f"Chunks created: {len(chunks)}\n"
            f"Stored in ChromaDB."
        )

    except Exception as e:
        return f"Error while processing video: {str(e)}"
```

```python
def process_video_or_transcript_for_ui(video_url, pasted_transcript):

    if pasted_transcript and pasted_transcript.strip():
        return process_pasted_transcript_for_ui(pasted_transcript)

    if video_url and video_url.strip():
        return process_video_for_ui(video_url)

    return "Please provide either a YouTube URL or paste a transcript."
```

```python
def process_pasted_transcript_for_ui(pasted_transcript):

    global CURRENT_VIDEO_ID

    try:
        CURRENT_VIDEO_ID = "pasted_transcript"

        transcript_segments = []

        words = pasted_transcript.split()
        chunk_size = 120

        start_time = 0

        for i in range(0, len(words), chunk_size):
            text = " ".join(words[i:i + chunk_size])

            segment = TranscriptSegment(
                video_id=CURRENT_VIDEO_ID,
                start_time=start_time,
                duration=60,
                text=text
            )

            transcript_segments.append(segment)
            start_time += 60

        chunks = create_chunks(transcript_segments)

        store_chunks(chunks)

        return f"""
Transcript processed successfully.

Source: Pasted transcript
Transcript segments created: {len(transcript_segments)}
Chunks created: {len(chunks)}

You can now use Video Roadmap, Learn, Study Pack, and Quiz.
"""

    except Exception as e:
        return f"Error while processing pasted transcript: {e}"
```

```python
def topics_for_ui():

    try:
        topic_list = extract_topics_structured()

        output = "# Topic Timeline\n\n"

        for topic in topic_list.topics:
            output += f"### {topic.start_time} - {topic.title}\n"
            output += f"{topic.description}\n\n"

        return output

    except Exception as e:
        return f"Error while extracting topics: {str(e)}"
```

```python
def study_pack_for_ui(topic_name):

    try:
        study_pack = generate_study_pack(topic_name)

        if study_pack is None:
            return "I could not find enough transcript content for this topic."

        output = f"""
# Study Pack: {study_pack.topic}

## Explanation

{study_pack.explanation}

## Key Points

"""

        for point in study_pack.key_points:
            output += f"- {point}\n"

        output += f"""

## Example

{study_pack.example}

## Interview-Style Summary

{study_pack.interview_style_summary}
"""

        return output

    except Exception as e:
        return f"Error while generating study pack: {str(e)}"
```

```python
def quiz_for_ui(topic_name):

    try:
        quiz = generate_quiz(topic_name)

        if quiz is None:
            return "I could not find enough transcript content for this quiz topic."

        output = f"# Quiz: {topic_name}\n\n"

        for i, question in enumerate(quiz.questions, start=1):

            output += f"## Question {i}\n"
            output += f"{question.question}\n\n"

            for option in question.options:
                output += f"- {option}\n"

            output += f"\n**Answer:** {question.answer}\n\n"
            output += f"**Explanation:** {question.explanation}\n\n"

        return output

    except Exception as e:
        return f"Error while generating quiz: {str(e)}"
```

# Step 12: Gradio UI

```python
with gr.Blocks() as demo:

    gr.Markdown(
        """
        # YouTube Study Companion

        A RAG-powered learning assistant for long educational YouTube videos.

        This app can process a YouTube transcript, store it in a vector database,
        retrieve relevant sections, answer questions, generate a topic timeline, create study packs, and build quizzes.
        """
    )

    with gr.Tab("Process Video"):

        video_url_input = gr.Textbox(
            label="YouTube Video URL",
            placeholder="Paste a YouTube video URL here"
        )

        transcript_input = gr.Textbox(
            label="Or Paste Transcript",
            placeholder="Paste transcript text here if YouTube transcript extraction is blocked",
            lines=10
        )

        process_button = gr.Button("Process Video / Transcript")

        process_status = gr.Textbox(
            label="Status",
            lines=5
        )

        process_button.click(
            process_video_or_transcript_for_ui,
            inputs=[video_url_input, transcript_input],
            outputs=process_status
        )

    with gr.Tab("Video Roadmap"):

        topics_button = gr.Button("Generate Topic Timeline")

        topics_output = gr.Markdown(
            label="Topic Timeline"
        )

        topics_button.click(
            topics_for_ui,
            outputs=topics_output
        )

    with gr.Tab("Learn"):

        question_input = gr.Textbox(
            label="Ask a question about the video",
            placeholder="Example: What is a benchmark?"
        )

        answer_button = gr.Button("Get Answer")

        answer_output = gr.Markdown(
            label="Answer"
        )

        answer_button.click(
            answer_question,
            inputs=question_input,
            outputs=answer_output
        )

    

    with gr.Tab("Study Pack"):

        study_topic_input = gr.Textbox(
            label="Topic Name",
            placeholder="Example: benchmarks, inner join, execution plan"
        )

        study_pack_button = gr.Button(
            "Generate Study Pack"
        )

        study_pack_output = gr.Markdown()

        study_pack_button.click(
            study_pack_for_ui,
            inputs=study_topic_input,
            outputs=study_pack_output
        )

    with gr.Tab("Quiz"):

        quiz_topic_input = gr.Textbox(
            label="Quiz Topic",
            placeholder="Example: benchmarks, inner join, execution plan"
        )

        quiz_button = gr.Button(
            "Generate Quiz"
        )

        quiz_output = gr.Markdown()

        quiz_button.click(
            quiz_for_ui,
            inputs=quiz_topic_input,
            outputs=quiz_output
        )
```

```python
demo.launch()
```
